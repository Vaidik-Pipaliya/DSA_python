```python
num = 24
if num%2==0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
#now we will make function
```

```python
def even_odd(num):
    """this will finds even or odd nummbers"""
    if num%2==0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
```

```python
##calling function
even_odd(5)
```

```python
##variable length arguments(*)
#postional arguments

def print_num(*args):
    for num in args:
        print(num)
print_num(1,2,3,4,5,"vaidik")
```

```python
# keyword arguments(**), key-value pair of parameters
def print_detail(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
print_detail(name="vd",agg=32,city="pune")
```

```python
#combining both
def print_details(*args,**kwargs):
    for val in args:
        print(f" positional args:-{val}")

    for key,value in kwargs.items():
        print(f"key word args:-{key}:{value}")

print_details(1,2,3,4,5,"vd",name="vd",agg=32,city="pune")
```
