# Problem Explanation

Checks if a hardcoded number `num` is even or odd and prints the result.

# Approach Used

* **Technique:** Simple Conditional check.
* **Why it works:** If a number is divisible by 2 with no remainder, it is even. otherwise, it is odd.
* **Key Idea:** Modulo division by 2.

# Dry Run

Input: `num = 24`
* `num % 2 == 0` -> `24 % 2` is `0`.
* Condition is `True`.
* Prints: `24 is even`.

# Time Complexity

* **Time:** O(1)
* **Explanation:** Execution takes constant time.

# Space Complexity

* **Space:** O(1)
* **Explanation:** No extra variables or data structures.


```python
def even_odd(num):
    """this will finds even or odd nummbers"""
    if num%2==0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
```

* **Key Idea:** Parameterized code execution.

```python
##variable length arguments(*)
#postional arguments

def print_num(*args):
    for num in args:
        print(num)
print_num(1,2,3,4,5,"vaidik")
```

# Problem Explanation

Defines a function `print_num` that accepts a variable number of positional arguments using `*args` and prints them one by one.

# Approach Used

* **Technique:** Variable length arguments (`*args`).
* **Why it works:** In Python, `*args` packs all positional arguments into a tuple, which can then be iterated over.
* **Key Idea:** Collect a variable number of positional parameters.

# Dry Run

Input: `print_num(1, 2, 3, 4, 5, "vaidik")`
* `args` becomes tuple: `(1, 2, 3, 4, 5, "vaidik")`
* Iterates through the tuple and prints each item.

# Time Complexity

* **Time:** O(M)
* **Explanation:** Where `M` is the number of arguments passed to the function.

# Space Complexity

* **Space:** O(M)
* **Explanation:** To store the tuple of arguments.


```python
# keyword arguments(**), key-value pair of parameters
def print_detail(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
print_detail(name="vd",agg=32,city="pune")
```

# Problem Explanation

Defines a function `print_detail` that accepts keyword arguments using `**kwargs` and prints them as key-value pairs.

# Approach Used

* **Technique:** Variable keyword arguments (`**kwargs`).
* **Why it works:** `**kwargs` packs keyword arguments into a dictionary. We iterate over key-value pairs using `.items()`.
* **Key Idea:** Collect dictionary-like variable parameters.

# Dry Run

Input: `print_detail(name="vd", agg=32, city="pune")`
* `kwargs` becomes: `{'name': 'vd', 'agg': 32, 'city': 'pune'}`
* Loop outputs key-value pairs.

# Time Complexity

* **Time:** O(K)
* **Explanation:** Where `K` is the number of keyword arguments.

# Space Complexity

* **Space:** O(K)
* **Explanation:** To store the keyword argument dictionary.


```python
#combining both
def print_details(*args,**kwargs):
    for val in args:
        print(f" positional args:-{val}")

    for key,value in kwargs.items():
        print(f"key word args:-{key}:{value}")

print_details(1,2,3,4,5,"vd",name="vd",agg=32,city="pune")
```

# Problem Explanation

Defines and calls a function `print_details` that combines both positional (`*args`) and keyword (`**kwargs`) variable arguments.

# Approach Used

* **Technique:** Combined `*args` and `**kwargs`.
* **Why it works:** Positional parameters are captured in `args` tuple, while named/keyword parameters are captured in `kwargs` dict.
* **Key Idea:** Full argument list collection.

# Dry Run

Input: `print_details(1, 2, 3, 4, 5, "vd", name="vd", agg=32, city="pune")`
* `args` tuple: `(1, 2, 3, 4, 5, "vd")`
* `kwargs` dict: `{'name': 'vd', 'agg': 32, 'city': 'pune'}`
* Loops print elements of `args` then items of `kwargs`.

# Time Complexity

* **Time:** O(M + K)
* **Explanation:** Proportional to the total number of arguments.

# Space Complexity

* **Space:** O(M + K)
* **Explanation:** To store the tuple and dictionary objects.
