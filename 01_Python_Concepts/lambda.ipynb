# Approach: Standard function definition

## Problem Explanation

Defines a standard sum function and calls it using input from the console.

## Approach Used

* **Technique:** Standard function definition.
* **Why it works:** Encapsulates addition operation using positional parameters `a` and `b`.
* **Key Idea:** Function mapping.

## Dry Run

Input: `a = 5`, `b = 6`
* Returns `5 + 6 = 11`.

# Approach: Lambda Expression

## Problem Explanation

Demonstrates how to define and use an anonymous lambda function to compute the sum of two numbers.

## Approach Used

* **Technique:** Lambda Expression.
* **Why it works:** Evaluates the expression `a + b` implicitly and returns the result.
* **Key Idea:** Anonymous in-line function.

## Dry Run

Input: `addition(5, 6)`
* Evaluates `5 + 6`.
* Returns `11`.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; Constant time evaluation.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; No extra variables.
</div>

```python
def even(num):
    if num % 2 == 0:
        print("num is even")
    else:
        print("num is odd")
```

# Approach: Conditional modulo check

## Problem Explanation

Defines a standard function `even` that prints if a number is even or odd.

## Approach Used

* **Technique:** Conditional modulo check.
* **Why it works:** Encapsulates the modulo 2 check in a named block.
* **Key Idea:** Even/odd check.

## Dry Run

Input: `even(4)`
* `4 % 2 == 0` is `True`.
* Prints `"num is even"`.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; Evaluates conditions in constant time.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; Constant memory.
</div>

```python
even = lambda a:a%2 == 0
print((even(1)))
```

# Approach: Lambda expression with boolean output

## Problem Explanation

Defines a lambda function that checks if a number is even, returning a boolean.

## Approach Used

* **Technique:** Lambda expression with boolean output.
* **Why it works:** Computes `a % 2 == 0` directly and returns the boolean value.
* **Key Idea:** Anonymous predicate.

## Dry Run

Input: `even(1)`
* Evaluates `1 % 2 == 0` which is `False`.
