# Problem Explanation

Defines a standard sum function and calls it using input from the console.

# Approach Used

* **Technique:** Standard function definition.
* **Why it works:** Encapsulates addition operation using positional parameters `a` and `b`.
* **Key Idea:** Function mapping.

# Dry Run

Input: `a = 5`, `b = 6`
* Returns `5 + 6 = 11`.

# Time Complexity

* **Time:** O(1)
* **Explanation:** Arithmetic operation takes constant time.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Local scope variables.


```python
#using lambda
#defining
lambda a,b:a+b

#stiring in different variables
addition = lambda a,b : a+b
print(addition(5,6))
```

# Problem Explanation

Demonstrates how to define and use an anonymous lambda function to compute the sum of two numbers.

# Approach Used

* **Technique:** Lambda Expression.
* **Why it works:** Evaluates the expression `a + b` implicitly and returns the result.
* **Key Idea:** Anonymous in-line function.

# Dry Run

Input: `addition(5, 6)`
* Evaluates `5 + 6`.
* Returns `11`.

# Time Complexity

* **Time:** O(1)
* **Explanation:** Constant time evaluation.

# Space Complexity

* **Space:** O(1)
* **Explanation:** No extra variables.


```python
def even(num):
    if num % 2 == 0:
        print("num is even")
    else:
        print("num is odd")
```

# Problem Explanation

Defines a standard function `even` that prints if a number is even or odd.

# Approach Used

* **Technique:** Conditional modulo check.
* **Why it works:** Encapsulates the modulo 2 check in a named block.
* **Key Idea:** Even/odd check.

# Dry Run

Input: `even(4)`
* `4 % 2 == 0` is `True`.
* Prints `"num is even"`.

# Time Complexity

* **Time:** O(1)
* **Explanation:** Evaluates conditions in constant time.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Constant memory.


```python
even = lambda a:a%2 == 0
print((even(1)))
```

# Approach Used
