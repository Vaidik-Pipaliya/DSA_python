# Problem Explanation

Applies a lambda function that squares each number to a list of numbers using `map()`.

# Approach Used

* **Technique:** `map` with lambda.
* **Why it works:** `map` applies the squaring lambda function to each item of the collection iteratively.
* **Key Idea:** Mapping square transformation.

* **Explanation:** Maps elements in linear sequence.
