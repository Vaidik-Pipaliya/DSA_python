```python
def func(x,n):
    if n == 0:
        return 
    print(x)
    func(x,n-1)
func(15,3)
```

# Problem Explanation

Recursively prints a value `x`, `n` times.

# Approach Used

* **Technique:** Recursion.
* **Why it works:** Decrements count `n` on each recursive call, printing `x` and returning when `n` reaches `0` (base case).
* **Key Idea:** Base-case driven recursive stack execution.

# Dry Run

Input: `func(15, 3)`
* `n = 3`: Prints `15`. Calls `func(15, 2)`.
* `n = 2`: Prints `15`. Calls `func(15, 1)`.
* `n = 1`: Prints `15`. Calls `func(15, 0)`.
* `n = 0`: Returns immediately.
* Recursion completes.

# Time Complexity

* **Time:** O(n)
* **Explanation:** The function runs exactly `n` recursive steps.

# Space Complexity

* **Space:** O(n)
* **Explanation:** Utilizes `n` stack frames for the recursion.
