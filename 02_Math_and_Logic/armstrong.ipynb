```python
def armstrong(n): #321
    num = n 
    total = 0
    length = len(str(n))
    while num > 0:
        ld = num % 10
        total = total+(ld ** length)
        num = num // 10
    return total == n

armstrong(321)
```

# Approach: Mathematical digit extraction

## Problem Explanation

Checks if a number is an Armstrong number (where the sum of its digits raised to the power of the total number of digits equals the original number).

## Approach Used

* **Technique:** Mathematical digit extraction.
* **Why it works:** Extracts digits one by one using `% 10`, raises each to the power of the length of the number, sums them, and compares the final sum to the original number.
* **Key Idea:** Digit extraction and power sum.

## Dry Run

Input: `armstrong(321)`
* `length = len("321") = 3`, `num = 321`, `total = 0`
* Iteration 1:
  * `ld = 321 % 10 = 1`
  * `total = 0 + 1**3 = 1`
  * `num = 321 // 10 = 32`
* Iteration 2:
  * `ld = 32 % 10 = 2`
  * `total = 1 + 2**3 = 1 + 8 = 9`
  * `num = 32 // 10 = 3`
* Iteration 3:
  * `ld = 3 % 10 = 3`
  * `total = 9 + 3**3 = 9 + 27 = 36`
  * `num = 3 // 10 = 0`
* Loop terminates. Checks if `total == n` (`36 == 321`), returning `False`.
