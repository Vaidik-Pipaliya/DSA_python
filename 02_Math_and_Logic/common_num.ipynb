# Brute Force

```python
n = [1,2,3,3,6,7,4,3,2,2,1,1,9,9,8,7]
m = [1,3,5,6,7,8,9,0,]

for i in m:
    count = 0
    for j in n:
        if  j == i:
            count+=1
    print(count)
```

# Approach: Brute Force Nested Loops

## Problem Explanation

Counts the occurrences of elements from list `m` in list `n` using brute force double loops.

## Approach Used

* **Technique:** Brute Force Nested Loops.
* **Why it works:** For each element in `m`, it runs an inner loop over all elements in `n` to count occurrences.
* **Key Idea:** Element-by-element counting.

## Dry Run

Input: `n = [1, 2, 3, 3]`, `m = [1, 3]`
* Loop `i = 1`:
  * Scan `n`: `j=1` (match), `j=2`, `j=3`, `j=3`. `count` = 1. Prints `1`.
* Loop `i = 3`:
  * Scan `n`: `j=1`, `j=2`, `j=3` (match), `j=3` (match). `count` = 2. Prints `2`.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(M * N)</span> &mdash; Nested loops execute `len(m) * len(n)` times.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; Only uses constant iteration and count variables.
</div>

# optimal

```python
n=[5,3,2,2,1,5,5,7,5,10]
m=[10,111,1,9,5,67,2]

# constarints :-> 1<=n[i]<=10
hash_list = [0]*11
for num in n:
    hash_list[num]+=1
for num in m:
    if num<1 or num>10:
        print(0)
    else:
        
        print(hash_list[num])
```

# Approach: Hashing/Frequency Array

## Problem Explanation

Counts occurrences of elements from list `m` in list `n` using frequency hashing.

## Approach Used

* **Technique:** Hashing/Frequency Array.
* **Why it works:** Pre-calculates counts of elements in `n` in an array `hash_list` where the index represents the element value. Then, checks counts in O(1) time.
* **Key Idea:** Count collection frequency in a lookup table.

## Dry Run

Input: `n = [5, 3, 2, 2, 1, 5, 5, 7, 5, 10]`, `m = [10, 111, 2]`
* `hash_list` of size 11 initialized to 0.
* After populating counts from `n`: `hash_list = [0, 1, 2, 1, 0, 4, 0, 1, 0, 0, 1]`.
* Query loop for elements in `m`:
  * Query `10`: valid index -> prints `hash_list[10]` which is `1`.
  * Query `111`: out of range -> prints `0`.
  * Query `2`: valid index -> prints `hash_list[2]` which is `2`.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N + M)</span> &mdash; One loop of size `N` to build frequencies, and one loop of size `M` to query results.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(K)</span> &mdash; Uses a fixed size array of size `K` (11 in this case, based on constraint).
</div>
