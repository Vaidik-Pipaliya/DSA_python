```python

def count_digits(n):
    num = n 
    count = 0
    while num > 0:
        count += 1
        num = num // 10
    return count
print(count_digits(1234))
```

# Problem Explanation

Counts the number of digits in an integer by repeatedly dividing the number by 10.

# Approach Used

* **Technique:** Iterative division.
* **Why it works:** Dividing an integer by 10 drops its last digit. Counting divisions until the number becomes 0 yields the total digits.
* **Key Idea:** Loop division.

# Dry Run

Input: `n = 1234`
* `num = 1234`, `count = 0`
* Iteration 1: `count` becomes `1`, `num` becomes `123`
* Iteration 2: `count` becomes `2`, `num` becomes `12`
* Iteration 3: `count` becomes `3`, `num` becomes `1`
* Iteration 4: `count` becomes `4`, `num` becomes `0`
* Returns `4`.

# Time Complexity

* **Time:** O(log10(N))
* **Explanation:** The number of divisions is proportional to the base-10 logarithm of the number.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Constant auxiliary memory.

# Problem Explanation

Counts the number of digits in a positive integer using logarithmic properties.

# Approach Used

* **Technique:** Logarithmic digit count.
* **Why it works:** The floor of the base-10 logarithm of any positive integer `N` plus `1` represents its digit count mathematically.
* **Key Idea:** Compute `int(log10(num) + 1)` directly.

# Dry Run

Input: `num = 7894`
* `log10(7894)` is approximately `3.897`.
* `3.897 + 1 = 4.897`.
* `int(4.897) = 4`.
* Returns `4`.

# Time Complexity

* **Time:** O(1)
* **Explanation:** Execution relies on a single constant-time mathematical calculation.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Constant memory.
