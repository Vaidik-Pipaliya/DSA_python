```python
def func(n):
    if n == 1 or n == 0:
        return 1
    return n * func(n-1)
func(4)
```

# Problem Explanation

Calculates the factorial of a number using recursion.

# Approach Used

* **Technique:** Recursion.
* **Why it works:** Recursively breaks down calculation as `n * factorial(n - 1)` until the base case `n = 0` or `n = 1` is reached, returning `1`.
* **Key Idea:** Base case and recursive multiplication step.

# Dry Run

Input: `func(4)`
* `func(4)` calls `4 * func(3)`
* `func(3)` calls `3 * func(2)`
* `func(2)` calls `2 * func(1)`
* `func(1)` hit base case -> returns `1`
* Evaluation: `4 * (3 * (2 * 1)) = 24`.

# Time Complexity

* **Time:** O(N)
* **Explanation:** There are `N` recursive stack calls.

# Space Complexity

* **Space:** O(N)
* **Explanation:** Proportional to the recursive call stack depth.
