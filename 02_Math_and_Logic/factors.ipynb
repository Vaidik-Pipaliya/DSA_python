# Brute-Force methode

````python

def find_all_factorials(num):
    fact=[]
    for i in range(1,num+1):
        if num%i==0:
            fact.append(i)
    return fact
find_all_factorials(10)
````

# Better Solution

````python
def factors(num):
    result=[]
    for i in range(1,num//2+1):
        if num%i == 0:
            result.append(i)
    result.append(num)
    return result
factors(10)
````

# Square Root Method (Simple)

### Idea:-
Factors always come in **pairs**.

Example for `36`:
```
1 × 36
2 × 18
3 × 12
4 × 9
6 × 6
```
After `6` (√36), the same pairs repeat in reverse:
```
9 × 4
12 × 3
18 × 2
36 × 1
```
So, we only check numbers from **1 to √n**.
Whenever `i` is a factor, `n // i` is also a factor.

````python
import math
def fact(num):
    result=[]
    for i in range(1,int(math.sqrt(num))+1):
        if num%i == 0:
            result.append(i)
            if num//i != i:
                result.append(num//i)
    return sorted(result)
fact(16)
````
