# Brute-Force methode

````python

def find_all_factorials(num):
    fact=[]
    for i in range(1,num+1):
        if num%i==0:
            fact.append(i)
    return fact
find_all_factorials(10)
````

# Problem Explanation

Finds all factors of a number by scanning every integer from `1` to `num`.

# Approach Used

* **Technique:** Brute Force Iterative Trial Division.
* **Why it works:** A number `i` is a factor of `num` if `num % i == 0`. Checking all numbers up to `num` ensures we find all factors.
* **Key Idea:** Scan every number up to `num`.

# Dry Run

Input: `find_all_factorials(10)`
* `fact = []`
* Check `i = 1`: `10 % 1 == 0` (True) -> `fact = [1]`
* Check `i = 2`: `10 % 2 == 0` (True) -> `fact = [1, 2]`
* Check `i = 3` to `4`: `10 % i != 0` -> no action.
* Check `i = 5`: `10 % 5 == 0` (True) -> `fact = [1, 2, 5]`
* Check `i = 6` to `9`: `10 % i != 0` -> no action.
* Check `i = 10`: `10 % 10 == 0` (True) -> `fact = [1, 2, 5, 10]`
* Returns `[1, 2, 5, 10]`.

# Time Complexity

* **Time:** O(N)
* **Explanation:** The loop runs `N` times (where `N = num`).

# Space Complexity

* **Space:** O(F)
* **Explanation:** To store the list of factors (where `F` is the number of factors).

````python
def factors(num):
    result=[]
    for i in range(1,num//2+1):
        if num%i == 0:
            result.append(i)
    result.append(num)
    return result
factors(10)
````

# Problem Explanation

Finds factors of a number, optimized by only checking numbers up to `num // 2` and appending `num` at the end.

# Approach Used

* **Technique:** Half-scan Trial Division.
* **Why it works:** Aside from the number itself, a number cannot have any factors greater than half of its value.
* **Key Idea:** Loop boundary reduction.

# Dry Run

Input: `factors(10)`
* Loop range is `1` to `5` (`10 // 2` = 5):
  * `i = 1`: factor -> `result = [1]`
  * `i = 2`: factor -> `result = [1, 2]`
  * `i = 3`, `4`: not factors
  * `i = 5`: factor -> `result = [1, 2, 5]`
* Append `num = 10` -> `result = [1, 2, 5, 10]`
* Returns `[1, 2, 5, 10]`.

# Time Complexity

* **Time:** O(N)
* **Explanation:** The loop runs `N / 2` times, which is still linear O(N) complexity.

# Space Complexity

* **Space:** O(F)
* **Explanation:** Stores the list of factors.


# Square Root Method (Simple)

### Idea:-
Factors always come in **pairs**.

Example for `36`:
```
1 × 36
2 × 18
3 × 12
4 × 9
6 × 6
```
After `6` (√36), the same pairs repeat in reverse:
```
9 × 4
12 × 3
18 × 2
36 × 1
```
So, we only check numbers from **1 to √n**.
Whenever `i` is a factor, `n // i` is also a factor.

````python
import math
def fact(num):
    result=[]
    for i in range(1,int(math.sqrt(num))+1):
        if num%i == 0:
            result.append(i)
            if num//i != i:
                result.append(num//i)
    return sorted(result)
fact(16)
````

# Problem Explanation

Finds all factors of a number, highly optimized by looping only up to the square root of the number.

# Approach Used

* **Technique:** Square Root factor pairing.
* **Why it works:** Factors always come in pairs. If `i` is a factor of `num`, then `num // i` is also a factor. Hence, all unique factors are found by checking integers up to `sqrt(num)`.
* **Key Idea:** Pairing factors `i` and `num // i`.

# Dry Run

Input: `fact(16)`
* Loop `i` from `1` to `int(math.sqrt(16)) = 4`:
  * `i = 1`: factor -> `result = [1]`. Partner `16 // 1 = 16` -> `result = [1, 16]`
  * `i = 2`: factor -> `result = [1, 16, 2]`. Partner `16 // 2 = 8` -> `result = [1, 16, 2, 8]`
  * `i = 3`: not factor
  * `i = 4`: factor -> `result = [1, 16, 2, 8, 4]`. Partner `16 // 4 = 4` (equals `i`, skip appending partner).
* Returns sorted result: `[1, 2, 4, 8, 16]`.

# Time Complexity

* **Time:** O(sqrt(N))
* **Explanation:** The loop runs only up to `sqrt(N)` times. Sorting the factors takes O(F log F) where `F` is the number of factors, which is very small.

# Space Complexity

* **Space:** O(F)
* **Explanation:** Stores the output list of factors.
