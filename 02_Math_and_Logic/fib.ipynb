```python
class solution(object):
    def fib(self,n):
        if n  == 0  or n ==1:
            return n
        return self.fib(n-1) + self.fib(n-2)
obj = solution()
print(obj.fib(10))
```

# Problem Explanation

Calculates the Fibonacci number at index `n` using standard recursion.

# Approach Used

* **Technique:** Recursion.
* **Why it works:** Evaluates Fibonacci sequence recursively using the relation `F(n) = F(n-1) + F(n-2)` with base cases `F(0) = 0` and `F(1) = 1`.
* **Key Idea:** Branching recursion.

# Dry Run

Input: `fib(3)`
* `fib(3)` returns `fib(2) + fib(1)`
* `fib(2)` returns `fib(1) + fib(0)`
* `fib(1)` returns `1`, `fib(0)` returns `0`. `fib(2) = 1 + 0 = 1`.
* `fib(3) = 1 + 1 = 2`.

# Time Complexity

* **Time:** O(2^N)
* **Explanation:** Each call splits into two child calls, forming a recursion tree of exponential height.

# Space Complexity

* **Space:** O(N)
* **Explanation:** Depth of the recursive call stack is proportional to `N`.
