# Approach: Two Pointers

## Problem Explanation

Checks if a string is a palindrome using two pointers starting from opposite ends.

## Approach Used

* **Technique:** Two Pointers.
* **Why it works:** Compares letters at `left` and `right` indices. If they match, move the pointers closer until they cross. If any mismatch occurs, it's not a palindrome.
* **Key Idea:** Boundary comparison inward.

## Dry Run

Input: `s = "abccba"`
* `n = 6`, `left = 0`, `right = 5`
* `left = 0, right = 5`: `s[0] == s[5]` ('a' == 'a') -> `left = 1, right = 4`
* `left = 1, right = 4`: `s[1] == s[4]` ('b' == 'b') -> `left = 2, right = 3`
* `left = 2, right = 3`: `s[2] == s[3]` ('c' == 'c') -> `left = 3, right = 2`
* Loop exits as `left >= right`. Output: `True`.
