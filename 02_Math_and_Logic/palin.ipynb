### same from start and end both
e.g. :- NITIN, MOM

```python
s="abccba"
n = len(s)
left = 0
right = n-1
is_palin = True
while left<right:
    if s[left]!= s[right]:
        is_palin = False
        break
        
    left +=1
    right -= 1
print(is_palin)

```

```python
def func(s,right,left):
    if s[left]>=s[right]:
        return True
    if s[left]!=s[right]:
        return False
    return func(s,left+1,right-1)





s="abccba"
func(s,0,n-1)
```
