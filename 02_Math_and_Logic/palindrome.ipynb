```python
def palindrome(num):
    original = num
    result = 0

    while num > 0:
        last_digit = num % 10
        result = result * 10 + last_digit
        num = num // 10

    if original == result:
        return True
    else:
        return False

print(palindrome(121))
print(palindrome(-123))
```

# Problem Explanation

Checks if a positive integer is a palindrome by reversing its digits and comparing the result with the original number.

# Approach Used

* **Technique:** Digit Reversal.
* **Why it works:** Rebuilds the number backwards using modulo and division, then compares the reversed value with the original stored copy.
* **Key Idea:** Build numeric reverse.

# Dry Run

Input: `palindrome(121)`
* `original = 121`, `result = 0`, `num = 121`
* Iteration 1:
  * `last_digit = 121 % 10 = 1`
  * `result = 0 * 10 + 1 = 1`
  * `num = 121 // 10 = 12`
* Iteration 2:
  * `last_digit = 12 % 10 = 2`
  * `result = 1 * 10 + 2 = 12`
  * `num = 12 // 10 = 1`
* Iteration 3:
  * `last_digit = 1 % 10 = 1`
  * `result = 12 * 10 + 1 = 121`
  * `num = 1 // 10 = 0`
* Loop exits. `original == result` (`121 == 121`) is `True`. Returns `True`.

# Time Complexity

* **Time:** O(log10(N))
* **Explanation:** Loop runs once for each digit of the number.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Uses constant space.
