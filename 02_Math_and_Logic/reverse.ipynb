```python
n=5873
num = n
while num>0:
    last_digit= num%10
    print(last_digit)
    num = num//10
```

# Problem Explanation

Extracts and prints the digits of a positive integer in reverse order.

# Approach Used

* **Technique:** Modulo division digit extraction.
* **Why it works:** Taking `num % 10` extracts the last digit, which is printed. Division `num // 10` discards the last digit.
* **Key Idea:** Digit printing from right to left.

# Dry Run

Input: `n = 5873`
* `num = 5873`
* Iteration 1: `last_digit = 5873 % 10 = 3`. Prints `3`. `num` becomes `587`.
* Iteration 2: `last_digit = 587 % 10 = 7`. Prints `7`. `num` becomes `58`.
* Iteration 3: `last_digit = 58 % 10 = 8`. Prints `8`. `num` becomes `5`.
* Iteration 4: `last_digit = 5 % 10 = 5`. Prints `5`. `num` becomes `0`.
* Loop terminates.

# Time Complexity

* **Time:** O(log10(N))
* **Explanation:** Iterations are proportional to the number of digits in `N`.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Runs in-place with constant memory.
