```python
def fun(n):
    if n == 1:
        return 1
    return n + fun(n-1)
fun(5)
```

# Problem Explanation

Calculates the sum of first `n` natural numbers using recursion.

# Approach Used

* **Technique:** Recursion.
* **Why it works:** Recursively defines sum as `n + sum(n-1)` with the base case returning `1` when `n = 1`.
* **Key Idea:** Accumulation via call stack.

# Dry Run

Input: `fun(5)`
* `fun(5) = 5 + fun(4)`
* `fun(4) = 4 + fun(3)`
* `fun(3) = 3 + fun(2)`
* `fun(2) = 2 + fun(1)`
* `fun(1)` returns `1`.
* Reconstructs: `5 + 4 + 3 + 2 + 1 = 15`.

# Time Complexity

* **Time:** O(N)
* **Explanation:** Performs `N` recursive function calls.

# Space Complexity

* **Space:** O(N)
* **Explanation:** Stack frame depth is equal to `N`.
