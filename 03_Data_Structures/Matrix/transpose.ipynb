# changing row to columns and colums to rows



## transforming  3* 2 into 2 * 3



```python
nums = [[5,9,1],[2,3,7]]
print(f"actual matrix:-",nums)
row = len(nums)
col = len(nums[0])
result = [[0]*row for _ in range(col)]
for i in range(0,row):
    for j in range(0,col):
        result[j][i]=nums[i][j]
print(f"transposed matrix:-",result)
```
