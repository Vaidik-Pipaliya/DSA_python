```python
import array
arr = array.array('i',[1,2,3,4,5])
print(arr)
arr[0] = 3
print(arr)

#here "i" is not the element of array but it shows integer datatype of array
```

# Problem Explanation

Demonstrates initializing a type-restricted array using Python's standard `array` module and modifying its first element.

# Approach Used

* **Technique:** Standard array usage.
* **Why it works:** The module `array` defines space-efficient sequences of basic types. The parameter `'i'` specifies that elements must be signed integers.
* **Key Idea:** Type-constrained numerical array.

# Dry Run

Input: `arr = array('i', [1, 2, 3, 4, 5])`
* Prints: `array('i', [1, 2, 3, 4, 5])`
* `arr[0] = 3` modifies first index value.
* Prints: `array('i', [3, 2, 3, 4, 5])`

# Time Complexity

* **Time:** O(1)
* **Explanation:** Element assignment at index 0 and printing runs in constant time.

# Space Complexity

* **Space:** O(N)
* **Explanation:** To store the array elements in memory.
