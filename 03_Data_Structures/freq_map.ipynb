```python
def freq(num):
    dict = {}
    for i in range(0,len(num)):
        if num[i] in dict:
            dict[num[i]]+=1
        else:
            dict[num[i]]=1
    return dict
num = [1,1,2,3,4,4,5,7,8,8,6,9,9,4,4]
freq(num)
```

# Approach: Dictionary Hashing

## Problem Explanation

Counts the frequency of each element in a list using a dictionary with manual existence checks.

## Approach Used

* **Technique:** Dictionary Hashing.
* **Why it works:** Iterates through the list, checking if the element is already a key in the dictionary. If so, increments the value; else, initializes it to 1.
* **Key Idea:** Count elements using a lookup hash map.

## Dry Run

Input: `num = [1, 1, 2]`
* `dict = {}`
* `i = 0`: `num[0] = 1` -> not in dict -> `dict = {1: 1}`
* `i = 1`: `num[1] = 1` -> in dict -> `dict = {1: 2}`
* `i = 2`: `num[2] = 2` -> not in dict -> `dict = {1: 2, 2: 1}`
* Returns `{1: 2, 2: 1}`.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; Loops through the list of length `N` once. Dict lookups and insertions take O(1) average time.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(U)</span> &mdash; Where `U` is the number of unique elements stored in the dictionary.
</div>

# Methode 2 (using .get())

```python
num = [1,1,2,3,4,4,5,7,8,8,6,9,9,4,4]
def freq(num):
    h_map={}
    n=len(num)
    for i in range(0,n):
        h_map[num[i]] = h_map.get(num[i],0)+1
    return h_map
freq(num)
```

# Approach: Dictionary `.get()` method

## Problem Explanation

Counts the frequency of each element in a list using a dictionary with the `.get()` default value method.

## Approach Used

* **Technique:** Dictionary `.get()` method.
* **Why it works:** `h_map.get(key, default)` returns the current count of `key` if it exists, otherwise it returns the `default` value of `0`. We add `1` and reassign it.
* **Key Idea:** Simplified dictionary counting.

## Dry Run

Input: `num = [1, 1, 2]`
* `h_map = {}`, `n = 3`
* `i = 0`: `num[0] = 1` -> `h_map[1] = h_map.get(1, 0) + 1 = 0 + 1 = 1` -> `h_map = {1: 1}`
* `i = 1`: `num[1] = 1` -> `h_map[1] = h_map.get(1, 0) + 1 = 1 + 1 = 2` -> `h_map = {1: 2}`
* `i = 2`: `num[2] = 2` -> `h_map[2] = h_map.get(2, 0) + 1 = 0 + 1 = 1` -> `h_map = {1: 2, 2: 1}`
* Returns `{1: 2, 2: 1}`.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; Performs a single pass over the list of size `N`.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(U)</span> &mdash; Stores up to `U` unique elements.
</div>
