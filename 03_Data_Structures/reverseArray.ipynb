# Approach: Slicing and built-in methods

## Problem Explanation

Demonstrates reversing a list using built-in Python techniques: the `.reverse()` method and list slicing `[::-1]`.

## Approach Used

* **Technique:** Slicing and built-in methods.
* **Why it works:** `.reverse()` reverses the list in-place. `nums[::-1]` creates a reversed slice copy. Note that the expression `nums.reverse() and nums[::-1]` evaluates to `None` in python.
* **Key Idea:** Leverage built-in reversal mechanisms.

## Dry Run

Input: `nums = [5, 9, 8, 2, 3, 0, 6]`
* `nums.reverse()` mutates the list to `[6, 0, 3, 2, 8, 9, 5]`.
* The statement expression evaluates `None and slice` which resolves to `None`.
* Prints: `[6, 0, 3, 2, 8, 9, 5]`.

# using recursion

```python
def func (arr,left,right):
    if left >= right:
        return
    arr[left],arr[right] = arr[right],arr[left]
    func(arr,left+1,right-1)

def rev(nums):  
    func(nums,0,len(nums)-1)
    return nums
nums = [5,9,8,2,3,0,6]  
print(rev(nums))

   
```

# Approach: Recursion with Two Pointers

## Problem Explanation

Reverses an array recursively using two pointers swapping elements from the outer boundaries inward.

## Approach Used

* **Technique:** Recursion with Two Pointers.
* **Why it works:** Swaps elements at `left` and `right` indices, then calls itself recursively with `left + 1` and `right - 1` until the pointers meet or cross (`left >= right`).
* **Key Idea:** Sub-array reversal.

## Dry Run

Input: `nums = [5, 9, 8, 2]`
* `rev(nums)` calls `func(nums, 0, 3)`
* `left = 0`, `right = 3` (0 < 3):
  * Swaps `nums[0]` and `nums[3]` -> `[2, 9, 8, 5]`
  * Calls `func(nums, 1, 2)`
* `left = 1`, `right = 2` (1 < 2):
  * Swaps `nums[1]` and `nums[2]` -> `[2, 8, 9, 5]`
  * Calls `func(nums, 2, 1)`
* `left = 2`, `right = 1` (2 >= 1) -> base case hit, returns.
* Final Output: `[2, 8, 9, 5]`.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; Recursion depth is `N / 2`, and each step performs a constant-time swap.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; Recursive call stack depth is `N / 2`.
</div>
