```python
def bin_search(arr, target):

    start = 0
    end = len(arr)-1

    while start <= end:

        mid = (start + end) // 2

        if arr[mid] == target:
            print(f"Target found at index {mid}")
            return target

        elif arr[mid] > target:
            end = mid - 1

        else:
            start = mid + 1

    return -1


sort_list = [10,20,45,96,100]

target = 96

result = bin_search(sort_list, target)

print(result)
```

# Approach: Binary Search

## Problem Explanation

Implements Binary Search to locate a target element in a sorted list.

## Approach Used

* **Technique:** Binary Search.
* **Why it works:** In a sorted list, we can repeatedly divide the search interval in half. Compare the target to the middle element. If they match, return the index. If target is smaller, search the left half; if target is larger, search the right half.
* **Key Idea:** Logarithmic search space reduction.

## Dry Run

Input: `sort_list = [10, 20, 45, 96, 100]`, `target = 96`
* `start = 0`, `end = 4`
* Iteration 1:
  * `mid = (0 + 4) // 2 = 2`
  * `arr[2] = 45`. Since `45 < 96`, `start` becomes `mid + 1 = 3`.
* Iteration 2:
  * `mid = (3 + 4) // 2 = 3`
  * `arr[3] = 96`. Matches `target`.
  * Prints `"Target found at index 3"`, returns `96`.
