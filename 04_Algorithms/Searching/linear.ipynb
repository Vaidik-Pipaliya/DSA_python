```python
def linear_search(arr,target): #here we can use my_list as a parameter
    size = len(arr)
    for index in range(0,size):
        if arr[index] == target:
            return 0
    else:
        return -1



my_list=[10,20,30,40,50,100]
target = 500
result = linear_search(my_list,target)
print(result)
```
