```python
def linear_search(arr,target): #here we can use my_list as a parameter
    size = len(arr)
    for index in range(0,size):
        if arr[index] == target:
            return 0
    else:
        return -1



my_list=[10,20,30,40,50,100]
target = 500
result = linear_search(my_list,target)
print(result)
```

# Approach: Linear Search

## Problem Explanation

Performs a linear search to locate a target in an array, returning `0` if found and `-1` otherwise. Note that the implementation returns `0` instead of the matching index upon success.

## Approach Used

* **Technique:** Linear Search.
* **Why it works:** Iterates through every index of the array from start to end, checking if the element at that index equals the target.
* **Key Idea:** Sequential scan.

## Dry Run

Input: `my_list = [10, 20, 30]`, `target = 500`
* `size = 3`
* `index = 0`: `my_list[0] = 10 != 500`
* `index = 1`: `my_list[1] = 20 != 500`
* `index = 2`: `my_list[2] = 30 != 500`
* Loop completes without match, enters `else` block and returns `-1`.
