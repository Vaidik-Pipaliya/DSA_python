```python
def bubble_sort(arr):
    size = len(arr)
    for passes in range(0,size-1):
        swapped = False
        for j in range(0,size-1-passes):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1] =  arr[j+1],arr[j]
                swapped = True
        if not swapped:
            break
    return arr


unsorted_list = [8,1,9,3,8,11,12,5]
sorted_list = bubble_sort(unsorted_list)
print("sorted_list", sorted_list)
#Har pass ke baad ek element already apni correct position (end) par pahunch jata hai, isliye usko dobara check karne ki zarurat nahi hoti.
#Isliye size - passes - 1 likhte hain, taaki har baar unnecessary comparisons kam ho jaye aur algorithm fast ho jaye.
```

# Approach: Bubble Sort

## Problem Explanation

Sorts an array in ascending order using Bubble Sort with an early-stopping optimization.

## Approach Used

* **Technique:** Bubble Sort.
* **Why it works:** Repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order. The largest element "bubbles" to its correct position at the end in each pass. If a full pass completes without any swaps, the list is already sorted, and we terminate early.
* **Key Idea:** Adjacent comparisons and bubbling.

## Dry Run

Input: `arr = [3, 1, 2]`
* `size = 3`
* Pass `passes = 0`: `swapped = False`
  * `j = 0`: `arr[0] > arr[1]` (3 > 1) -> swap -> `arr = [1, 3, 2]`, `swapped = True`
  * `j = 1`: `arr[1] > arr[2]` (3 > 2) -> swap -> `arr = [1, 2, 3]`, `swapped = True`
* Pass `passes = 1`: `swapped = False`
  * `j = 0`: `arr[0] > arr[1]` (1 > 2) is False.
  * Loop for `j` ends. `swapped` remains `False`, break loop.
* Sorted Output: `[1, 2, 3]`.

## Time Complexity
