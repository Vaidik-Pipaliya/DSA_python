```python
def insertion_sort(arr):
    n=len(arr)
    for current in range(1,n):
        currentcard = arr[current]
        correctpos= current-1
        while correctpos >=0:
            if currentcard>arr[correctpos]:
                break
            else:
                arr[correctpos+1] = arr[correctpos]
                correctpos-=1
        arr[correctpos+1] =currentcard
    return arr


unsorted_list = [12,25,11,34,90,22]
sorted_list = insertion_sort(unsorted_list)
print(sorted_list)
```
