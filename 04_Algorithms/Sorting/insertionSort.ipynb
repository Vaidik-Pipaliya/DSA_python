```python
def insertion_sort(arr):
    n=len(arr)
    for current in range(1,n):
        currentcard = arr[current]
        correctpos= current-1
        while correctpos >=0:
            if currentcard>arr[correctpos]:
                break
            else:
                arr[correctpos+1] = arr[correctpos]
                correctpos-=1
        arr[correctpos+1] =currentcard
    return arr


unsorted_list = [12,25,11,34,90,22]
sorted_list = insertion_sort(unsorted_list)
print(sorted_list)
```

# Problem Explanation

Sorts an array in ascending order using the Insertion Sort algorithm.

# Approach Used

* **Technique:** Insertion Sort.
* **Why it works:** Divides the array into sorted and unsorted parts. For each element in the unsorted part, it is inserted into its correct position in the sorted prefix by shifting larger elements one position to the right.
* **Key Idea:** Shift and insert.

# Dry Run

Input: `arr = [3, 1, 2]`
* `n = 3`
* `current = 1`: `currentcard = 1`, `correctpos = 0`
  * Loop condition `correctpos >= 0` and `currentcard <= arr[0]` (1 < 3):
    * `arr[1] = arr[0]` (3) -> `arr = [3, 3, 2]`
    * `correctpos = -1`
  * Exit loop. `arr[0] = currentcard` (1) -> `arr = [1, 3, 2]`
* `current = 2`: `currentcard = 2`, `correctpos = 1`
  * Loop condition `correctpos >= 0` and `currentcard <= arr[1]` (2 < 3):
    * `arr[2] = arr[1]` (3) -> `arr = [1, 3, 3]`
    * `correctpos = 0`
  * Loop condition `correctpos >= 0` and `currentcard > arr[0]` (2 > 1) -> breaks.
  * `arr[1] = currentcard` (2) -> `arr = [1, 2, 3]`
* Sorted Output: `[1, 2, 3]`.

# Time Complexity

* **Time:** O(N^2)
* **Explanation:** Outer loop runs `N` times; inner loop can shift up to `N` elements in the worst case (reverse sorted list).

# Space Complexity

* **Space:** O(1)
* **Explanation:** Performs sorting in-place.
