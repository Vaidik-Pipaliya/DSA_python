```python
#merging part after sorting
def merg(left,right):
    result=[]
    i,j=0,0
    n,m=len(left),len(right)
    while i<n and j<m:
        if left[i]<=right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
#checking if any array is exausted or not!!
    if i<n:
        while i<n:
            result.append(left[i])
            i+=1
    if j<m:
        while j<m:
            result.append(right[j])
            j+=1
    return result

#merge sort (recursion will be used and for recursion we muts have one base case )
def merge_sort(arr):
    if len(arr)<=1:
        return arr
    mid = len(arr)//2
    left_arr = arr[:mid]
    right_arr = arr[mid:]
    left =merge_sort(left_arr)
    right = merge_sort(right_arr)

    return merg(left,right)

nums = [3,1,2,4,1,5,2,6,4]
merge_sort(nums)
```

# Approach: Merge Sort (Divide and Conquer)

## Problem Explanation

Sorts an array in ascending order using the Divide and Conquer Merge Sort algorithm.

## Approach Used

* **Technique:** Merge Sort (Divide and Conquer).
* **Why it works:** Recursively splits the array into two halves until subarrays contain at most one element (which are naturally sorted). Then, merges sorted halves back together in sorted order.
* **Key Idea:** Divide array in halves, sort recursively, and merge.

## Dry Run

Input: `nums = [3, 1, 2]`
* `merge_sort([3, 1, 2])`:
  * Split to `left_arr = [3]`, `right_arr = [1, 2]`
  * `merge_sort([3])` returns `[3]`
  * `merge_sort([1, 2])`:
    * Split to `left_arr = [1]`, `right_arr = [2]`
    * Merging `[1]` and `[2]` -> returns `[1, 2]`
  * Merge `left = [3]` and `right = [1, 2]`:
    * Compare `3` and `1` -> result `[1]`
    * Compare `3` and `2` -> result `[1, 2]`
    * Exhausted right array, append remaining left `[3]` -> result `[1, 2, 3]`.
* Output: `[1, 2, 3]`.
