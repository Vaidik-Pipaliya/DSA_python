```python
def selection_sort(num):
    n=len(num)
    for i in range(0,num):
        min_index = i
        for j in range(i+1,n):
            if num[j]<num[min_index]:
                min_index = j
        num[i],num[min_index] = num[min_index],num[i]
```

# Problem Explanation

Sorts an array in ascending order using Selection Sort. Note that this implementation contains a bug in the outer loop range statement (`range(0, num)` where `num` is the input list, raising a `TypeError`).

# Approach Used

* **Technique:** Selection Sort (Buggy implementation).
* **Why it works:** In each pass, it attempts to find the minimum element in the unsorted suffix and swap it with the element at the current boundaries index.
* **Key Idea:** Select minimum and swap.

# Time Complexity

* **Time:** O(N^2)
* **Explanation:** If fixed to `range(0, n)`, the nested comparisons check all pairs.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Operations are performed in-place.
