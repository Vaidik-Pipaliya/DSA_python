```python
item=["apple","banana","mango","graps"]
item.append("pitch")
for a in item:
    print("current items")
    print(f"-{a}")
```

# Approach: List operations

## Problem Explanation

Adds a fruit item to an inventory list and prints all the current items.

## Approach Used

* **Technique:** List operations.
* **Why it works:** Appends `'pitch'` to the list and uses a simple `for` loop to print each element.
* **Key Idea:** Inventory tracking.

## Dry Run

Input: `item = ["apple", "banana", "mango", "graps"]`
* `item.append("pitch")` modifies list to `["apple", "banana", "mango", "graps", "pitch"]`.
* Loop prints each item sequentially.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; Appending takes O(1) time. Printing loops through `N` elements taking O(N) time.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; No extra storage is created beyond the input list.
</div>

```python
item.remove("banana")
fruit = "mango"
if fruit in item:
    print(f"{fruit} is available")
else:
    print("not avilable")
print("inventory list:-")
for i in item:
    print(f"-{i}")
```

# Approach: Element removal and membership validation

## Problem Explanation

Removes a designated item from the inventory and verifies if another item exists in it.

## Approach Used

* **Technique:** Element removal and membership validation.
* **Why it works:** `.remove("banana")` searches and deletes the entry. The `in` operator evaluates if the target exists in the sequence.
* **Key Idea:** Membership testing and mutation.

## Dry Run

Input: `item = ["apple", "banana", "mango", "graps", "pitch"]`
* `item.remove("banana")` updates list to `["apple", "mango", "graps", "pitch"]`.
* `fruit = "mango"`. `"mango" in item` is `True`.
* Prints `"mango is available"`.
* Prints remaining list elements.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; `.remove()` and `in` search checks take linear time.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; Modifications are performed in-place.
</div>
