```python
# grades
grades=[50,60,70,80]
print("current grades:-",grades)
#adding grades
grades.append(40)
print("updated gardes:-",grades)

#average
average= sum(grades)/len(grades)
print(f"Average_grade:{average:.2f}")
```

# Approach: Aggregate arithmetic

## Problem Explanation

Appends a new score grade to a collection, and calculates the floating-point average grade formatted to two decimal places.

## Approach Used

* **Technique:** Aggregate arithmetic.
* **Why it works:** Appends item to list. Calculates sum using `sum()` and divides by number of items `len()`.
* **Key Idea:** Average calculation.

## Dry Run

Input: `grades = [50, 60, 70, 80]`
* `grades.append(40)` -> `[50, 60, 70, 80, 40]`.
* `sum(grades) = 300`, `len(grades) = 5`.
* `average = 300 / 5 = 60.0`.
* Prints `"Average_grade:60.00"`.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; `sum()` iterates through all `N` grades in linear time.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; Uses constant scalar variables.
</div>

```python
#finding minimum and maximum marks
maximum = max(grades)
minimum = min(grades)

print(f"highest grade:{maximum}")
print(f"lowest grade:{minimum}")
```

# Approach: Extrema search

## Problem Explanation

Finds the highest and lowest scores in a grades collection using Python's built-in `max()` and `min()` functions.

## Approach Used

* **Technique:** Extrema search.
* **Why it works:** Built-in functions scan the array to extract the maximum and minimum values.
* **Key Idea:** Max/min retrieval.

## Dry Run

Input: `grades = [50, 60, 70, 80, 40]`
* `maximum = max(grades) = 80`.
* `minimum = min(grades) = 40`.
* Prints: `highest grade:80`, `lowest grade:40`.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; Both `max()` and `min()` make a linear scan over `N` elements.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; Auxiliary memory is constant.
</div>
