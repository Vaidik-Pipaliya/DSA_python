```python
to_do_list=["cooking","billing","pay bills","clean car"] #pending work
print("current pending works",to_do_list)
```

# Problem Explanation

Initializes a list representing pending tasks and prints it to the console.

# Approach Used

* **Technique:** List instantiation.
* **Why it works:** Stores task strings in a list structure.
* **Key Idea:** Pending task tracking.

# Dry Run

* Prints: `current pending works ['cooking', 'billing', 'pay bills', 'clean car']`.

# Time Complexity

* **Time:** O(N)
* **Explanation:** Instantiation and printing take linear time.

# Space Complexity

* **Space:** O(N)
* **Explanation:** Storage for N task strings.


```python
#adding work
to_do_list.append("washing bike")
to_do_list.append("washing dishes")
print("all pennding works of today:-",to_do_list)
```

# Problem Explanation

Adds new items to the task list using the `append()` method.

# Approach Used

* **Technique:** List insertion.
* **Why it works:** Dynamically appends task items to the end of the list.
* **Key Idea:** Task insertion.

# Dry Run

* Appends `"washing bike"` and `"washing dishes"`.
* Prints updated list.

# Time Complexity

* **Time:** O(1)
* **Explanation:** Appending to a list takes amortized constant time.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Constant extra space per append.


```python
#removing completed work
to_do_list.remove("cooking")
print("current pending works:",to_do_list)
```

# Problem Explanation

Removes a completed task from the to-do list using the `remove()` method.

# Approach Used

* **Technique:** Element deletion.
* **Why it works:** Searches the list for the matching string and removes it.
* **Key Idea:** Task completion removal.

# Dry Run

* `to_do_list.remove("cooking")` searches and deletes the entry.
* Prints updated list.

# Time Complexity

* **Time:** O(N)
* **Explanation:** Searching for the element to remove takes linear time.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Modified in-place.
