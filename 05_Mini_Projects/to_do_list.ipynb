```python
to_do_list=["cooking","billing","pay bills","clean car"] #pending work
print("current pending works",to_do_list)
```

# Approach: List instantiation

## Problem Explanation

Initializes a list representing pending tasks and prints it to the console.

## Approach Used

* **Technique:** List instantiation.
* **Why it works:** Stores task strings in a list structure.
* **Key Idea:** Pending task tracking.

## Dry Run

* Prints: `current pending works ['cooking', 'billing', 'pay bills', 'clean car']`.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; Instantiation and printing take linear time.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; Storage for N task strings.
</div>

```python
#adding work
to_do_list.append("washing bike")
to_do_list.append("washing dishes")
print("all pennding works of today:-",to_do_list)
```
