```python
feedback=["great work!","good work","it should be better"]
#adding negative
feedback.append("not happy with service")

positive_feedback_count= sum(1 for comment in feedback if "great" in comment.lower() or "good" in comment.lower())
print(f"positive feedback count: {positive_feedback_count}")

#printing all feedback
print("user feedback")
for comment in feedback:
    print(f"-{comment}")
```
