```python
feedback=["great work!","good work","it should be better"]
#adding negative
feedback.append("not happy with service")

positive_feedback_count= sum(1 for comment in feedback if "great" in comment.lower() or "good" in comment.lower())
print(f"positive feedback count: {positive_feedback_count}")

#printing all feedback
print("user feedback")
for comment in feedback:
    print(f"-{comment}")
```

# Approach: List traversal with string search

## Problem Explanation

Appends a comment to a user feedback list, counts positive comments (those containing the words "great" or "good"), and prints all feedbacks.

## Approach Used

* **Technique:** List traversal with string search.
* **Why it works:** Evaluates membership of substring tokens `"great"` or `"good"` inside lowercase versions of each comment, summing matching counts.
* **Key Idea:** Basic sentiment analysis counting.

## Dry Run

Input: `feedback = ["great work!", "good work", "it should be better", "not happy with service"]`
* Loop evaluation:
  * Comment 1: `"great work!"` contains `"great"` -> True.
  * Comment 2: `"good work"` contains `"good"` -> True.
  * Comment 3: `"it should be better"` contains neither -> False.
  * Comment 4: `"not happy with service"` contains neither -> False.
* `positive_feedback_count = 1 + 1 = 2`.
* Prints all feedback items.
