# BruteForce

```python
nums = [5,9,1,2,4,15,6,3]
def twosum(nums):
    n = len(nums)
    target = 13
    for i in range(0,n-1):
        for j in range(i+1,n):
            if nums[i]+nums[j] == target:
                return [i,j]

twosum(nums)            
```
