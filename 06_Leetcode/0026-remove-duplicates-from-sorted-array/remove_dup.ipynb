# brute-force

```python
num=[1,1,1,2,3,4,4,7,9,9,9,10]
def func(num):
    n=len(num)
    freq_map = {}
    for i in range(0,n):
        freq_map[num[i]] = 0
    j = 0
    for k in range(0,n):
        num[j]=k
        j+=1
    return j
```

# Approach: Index-based Overwriting (Faulty Logic)

## Problem Explanation

Intended to remove duplicates, but this implementation overwrites array elements with index counters `k` from `0` to `n-1`, returning the original length `n`.

## Approach Used

* **Technique:** Index-based Overwriting (Faulty Logic).
* **Why it works:** It populates a frequency map keys, but then immediately overwrites `num[j]` with numbers `0` to `n-1` without using the frequency map.
* **Key Idea:** Overwrite array entries with consecutive integers.

## Dry Run

Input: `num = [1, 1, 1, 2]`
* `n = 4`, `freq_map` keys become `{1: 0, 2: 0}`
* Loop over `k` from `0` to `3`:
  * `k = 0`: `num[0] = 0`, `j = 1`
  * `k = 1`: `num[1] = 1`, `j = 2`
  * `k = 2`: `num[2] = 2`, `j = 3`
  * `k = 3`: `num[3] = 3`, `j = 4`
* Returns `j = 4`. The array is mutated to `[0, 1, 2, 3]`.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; Runs two independent loops from `0` to `N-1`.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(U)</span> &mdash; Where `U` is the number of unique elements stored in `freq_map`.
</div>

# optimal

```python
num=[1,1,1,2,3,4,4,7,9,9,9,10]
def find(num):
    n = len(num)
    if n == 1:
        return 1
    i=0
    j=i+1
    while j<n:
        if num[j]!=num[i]:
            i+=1
            num[i],num[j] = num[j],num[i]
        j+=1
    return i+1 #total unique number jova mate
find(num)
```

# Approach: Two Pointers

## Problem Explanation

Removes duplicates from a sorted array in-place and returns the number of unique elements.

## Approach Used

* **Technique:** Two Pointers.
* **Why it works:** Uses pointer `i` to keep track of the position of the last unique element and pointer `j` to scan ahead. When a new element is found, `i` is incremented and swap is performed.
* **Key Idea:** Shift unique elements to the front of the array.

## Dry Run

Input: `num = [1, 1, 1, 2, 3]`
* `n = 5`, `i = 0`, `j = 1`
* `j = 1`: `num[1] == num[0]` (1 == 1) -> no action.
* `j = 2`: `num[2] == num[0]` (1 == 1) -> no action.
* `j = 3`: `num[3] != num[0]` (2 != 1) -> `i` becomes `1`. Swap `num[1]` and `num[3]`. Array becomes `[1, 2, 1, 1, 3]`.
* `j = 4`: `num[4] != num[1]` (3 != 2) -> `i` becomes `2`. Swap `num[2]` and `num[4]`. Array becomes `[1, 2, 3, 1, 1]`.
* Returns `i + 1 = 3`. Unique elements are at the first 3 indices.

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; Single pass from index `1` to `N-1`.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; Performed in-place; uses constant extra variables.
</div>
