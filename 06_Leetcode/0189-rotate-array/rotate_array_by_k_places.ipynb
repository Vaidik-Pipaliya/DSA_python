# brute force

```python
nums=[3,9,5,6,7,2]
k = int(input("enter the numer for rotation"))
for _ in range(0,k):
    e  = nums.pop()
    nums.insert(0,e)
```

# Approach: Repeated insertion at index 0

## Problem Explanation

Rotates the array elements to the right by `k` positions. The value `k` is input by the user.

## Approach Used

* **Technique:** Repeated insertion at index 0.
* **Why it works:** Popping the last element from the list and inserting it at the first index shifts all other elements to the right. Doing this `k` times results in a right rotation by `k` places.
* **Key Idea:** Pop from the end and insert at index 0, repeated `k` times.

## Dry Run

Input: `nums = [3, 9, 5, 6, 7, 2]`, `k = 2`
* Iteration 1:
  * `e = nums.pop() = 2` -> `nums` becomes `[3, 9, 5, 6, 7]`
  * `nums.insert(0, 2)` -> `nums` becomes `[2, 3, 9, 5, 6, 7]`
* Iteration 2:
  * `e = nums.pop() = 7` -> `nums` becomes `[2, 3, 9, 5, 6]`
  * `nums.insert(0, 7)` -> `nums` becomes `[7, 2, 3, 9, 5, 6]`

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #d9534f; background-color: #f2dede; color: #a94442; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #d9534f; font-weight: bold;">O(k * N)</span> &mdash; <code>pop()</code> is <code>O(1)</code> but <code>insert(0, e)</code> takes <code>O(N)</code> to shift all elements. This is repeated <code>k</code> times.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; The rotation is done in-place, requiring only constant auxiliary space.
</div>


# optimized  (k%n)

```python
def k_rotate(nums):
    k = int(input("enter the numer for rotation"))
    n = len(nums)
    rotation = k%n

    for _ in range(0,rotation):
        e = nums.pop()
        nums.insert(0,e)
    print (nums)
nums=[3,9,5,6,7,2]
k_rotate(nums)
```

# Approach: Modulo-Optimized Shift

## Problem Explanation

Rotates the array elements to the right by `k` positions, optimizing by taking `k % n` to skip redundant full rotations.

## Approach Used

* **Technique:** Modulo-Optimized Shift.
* **Why it works:** Since rotating an array of size `n` by `n` steps results in the original array, we only need to perform `k % n` actual shifts.
* **Key Idea:** Reduce the number of rotation steps using the modulo operator.

## Dry Run

Input: `nums = [3, 9, 5, 6, 7, 2]`, `k = 8`
* `n = 6`, `rotation = 8 % 6 = 2`
* Iteration 1:
  * `e = nums.pop() = 2` -> `nums` becomes `[3, 9, 5, 6, 7]`
  * `nums.insert(0, 2)` -> `nums` becomes `[2, 3, 9, 5, 6, 7]`
* Iteration 2:
  * `e = nums.pop() = 7` -> `nums` becomes `[2, 3, 9, 5, 6]`
  * `nums.insert(0, 7)` -> `nums` becomes `[7, 2, 3, 9, 5, 6]`

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #f0ad4e; background-color: #fcf8e3; color: #8a6d3b; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #f0ad4e; font-weight: bold;">O((k % N) * N)</span> &mdash; The loop runs $k \% N$ times, with each iteration performing an <code>O(N)</code> insertion.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; The list is updated in-place.
</div>

# Approach: Three-Step Reversal

## Problem Explanation

Rotates the array elements to the right by `k` positions in-place using array reversal, which achieves the optimal time complexity of <code>O(N)</code>.

## Approach Used

* **Technique:** Three-Step Reversal.
* **Algorithm:**
  1. Reduce `k` by taking `k %= n`.
  2. Reverse the last `k` elements: `reverse(nums, n - k, n - 1)`.
  3. Reverse the remaining `n - k` elements: `reverse(nums, 0, n - k - 1)`.
  4. Reverse the entire array: `reverse(nums, 0, n - 1)`.

## Dry Run

Input: `nums = [3, 9, 5, 6, 7, 2]`, `k = 2`
* `n = 6`, `k = 2`
* **Step 1: Reverse last `k` elements (`[7, 2]`)**
  * `nums` becomes `[3, 9, 5, 6, 2, 7]`
* **Step 2: Reverse first `n - k` elements (`[3, 9, 5, 6]`)**
  * `nums` becomes `[6, 5, 9, 3, 2, 7]`
* **Step 3: Reverse the entire array**
  * `nums` becomes `[7, 2, 3, 9, 5, 6]`

## Time Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Time Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(N)</span> &mdash; Reversing different parts of the array takes linear time proportional to the array size.
</div>

## Space Complexity

<div style="padding: 8px 12px; border-left: 4px solid #5cb85c; background-color: #dff0d8; color: #3c763d; margin: 10px 0;">
  <strong>Space Complexity:</strong> <span style="color: #5cb85c; font-weight: bold;">O(1)</span> &mdash; Done in-place without using extra array copies.
</div>


```python

def reverse(nums, left, right):
    while left < right:
        nums[left], nums[right] = nums[right], nums[left]
        left += 1
        right -= 1


nums = [3, 9, 5, 6, 7, 2]
k = 2

n = len(nums)
k %= n

reverse(nums, n - k, n - 1)
reverse(nums, 0, n - k - 1)
reverse(nums, 0, n - 1)

print(nums)
```
