# simple loop


```python
nums = [9,6,4,2,3,5,7,0,1]
def find(nums):
    n = len(nums)
    for i in range(0,n+1):
        if i not in nums:
            return i
print(find(nums))
```

# using dictionory

```python
#preparing dict
nums = [9,6,4,2,3,5,7,0,1]
def find(nums):

    n = len(nums)
    dict = {}
    for i in range(0,n+1):
        dict[i]=0
    for num in nums:
        dict[num] =1
    for k,v in dict.items():
        if v == 0:
            return k
find(nums)
```
