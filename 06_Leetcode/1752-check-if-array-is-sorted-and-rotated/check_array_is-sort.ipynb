```python
nums = [3,5,6,8,9,10,20]
def check(nums):
    n = len(nums)
    for i in range(0,n-1):
        if nums[i]<nums[i+1]:
            return True
        else:
            return False
check(nums)
```

# Approach: Adjacent comparison with early return

## Problem Explanation

Checks if a list of numbers is sorted in ascending order. Note that the implementation only checks if the first element is less than the second and immediately returns, rather than scanning the entire list.

## Approach Used

* **Technique:** Adjacent comparison with early return.
* **Why it works:** It compares `nums[i]` and `nums[i+1]`. However, because it returns `True` on the first match and `False` on the first mismatch, it executes only for `i = 0` and then terminates.
* **Key Idea:** Compare `nums[0]` with `nums[1]` and return immediately.

## Dry Run

Input: `nums = [3, 5, 6, 8, 9, 10, 20]`
* `n = 7`
* Loop starts at `i = 0`:
  * Compares `nums[0] < nums[1]` which is `3 < 5` (evaluates to `True`).
  * Executes `return True` inside the `if` block.
* Function exits and returns `True`.
