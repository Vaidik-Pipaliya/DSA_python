take a new result with all zeros and two pointeers for negative and positive![alt text](image-1.png) 
then do p+2 for positves


```python
#optimal solutions using TWO POINTERS
nums = [5,10,-3,-1,-10,6]

def arrange(nums):
    n = len(nums)
    result = [0]*n
    pos  = 0
    neg = 1
    for i in range(0,n):
        if nums[i]>=0:
            result[pos] = nums[i]
            pos+=2
        else:
            result[neg] = nums[i]
            neg+=2
    return result
arrange(nums)
```
