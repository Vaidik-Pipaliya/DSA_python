```python
def bin_search(arr, target):

    start = 0
    end = len(arr)-1

    while start <= end:

        mid = (start + end) // 2

        if arr[mid] == target:
            print(f"Target found at index {mid}")
            return target

        elif arr[mid] > target:
            end = mid - 1

        else:
            start = mid + 1

    return -1


sort_list = [10,20,45,96,100]

target = 96

result = bin_search(sort_list, target)

print(result)
```
