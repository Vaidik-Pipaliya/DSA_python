# Brute Force

```python
n = [1,2,3,3,6,7,4,3,2,2,1,1,9,9,8,7]
m = [1,3,5,6,7,8,9,0,]

for i in m:
    count = 0
    for j in n:
        if  j == i:
            count+=1
    print(count)
```
