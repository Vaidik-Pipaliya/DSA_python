```python
def freq(num):
    dict = {}
    for i in range(0,len(num)):
        if num[i] in dict:
            dict[num[i]]+=1
        else:
            dict[num[i]]=1
    return dict
num = [1,1,2,3,4,4,5,7,8,8,6,9,9,4,4]
freq(num)
```

# Methode 2 (using .get())

```python
num = [1,1,2,3,4,4,5,7,8,8,6,9,9,4,4]
def freq(num):
    h_map={}
    n=len(num)
    for i in range(0,n):
        h_map[num[i]] = h_map.get(num[i],0)+1
    return h_map
freq(num)
```
