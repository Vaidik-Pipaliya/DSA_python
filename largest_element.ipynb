```python
#method 1
arr = [55,32,-97,99,3,67]
def large(arr):
    n = len(arr)
    largest = arr[0]
    for i in range(0,n):
        largest = max(largest,arr[i])
    return largest
large(arr)
```

# Problem Explanation

Finds the maximum element in an array by iteratively keeping track of the largest value using Python's built-in `max()` function.

# Approach Used

* **Technique:** Iterative Comparison using `max()`.
* **Why it works:** It initializes `largest` to the first element and updates it by taking the maximum of `largest` and the current element as it loops through the array.
* **Key Idea:** Iterate through each element and update the running maximum.

# Dry Run

Input: `arr = [55, 32, -97, 99, 3, 67]`
* `n = 6`, `largest = 55`
* Loop index `i = 0`: `largest = max(55, arr[0]) = max(55, 55) = 55`
* Loop index `i = 1`: `largest = max(55, arr[1]) = max(55, 32) = 55`
* Loop index `i = 2`: `largest = max(55, arr[2]) = max(55, -97) = 55`
* Loop index `i = 3`: `largest = max(55, arr[3]) = max(55, 99) = 99`
* Loop index `i = 4`: `largest = max(99, arr[4]) = max(99, 3) = 99`
* Loop index `i = 5`: `largest = max(99, arr[5]) = max(99, 67) = 99`
* Returns `99`.

# Time Complexity

* **Time:** O(N)
* **Explanation:** The loop runs exactly `N` times (where `N` is the length of the array).

# Space Complexity

* **Space:** O(1)
* **Explanation:** Only uses constant extra space for variables `largest` and `n`.


```python
#method 2
arr = [55,32,-97,99,3,67]
def large(arr):
    n = len(arr)
    largest = arr[0]
    for i in range(0,n):
        if arr[i]>largest:
            largest = arr[i]
    return largest
large(arr)
```

# Problem Explanation

Finds the maximum element in an array by iterating through the array and using an explicit conditional check (`if` statement) to update the largest value.

# Approach Used

* **Technique:** Iterative Comparison.
* **Why it works:** Traversing the array element-by-element guarantees we inspect every number, updating our candidate if a larger one is found.
* **Key Idea:** Compare the current element with the running largest value and update if the current element is greater.

# Dry Run

Input: `arr = [55, 32, -97, 99, 3, 67]`
* `n = 6`, `largest = 55`
* Loop index `i = 0`: `arr[0] > largest` (55 > 55) is `False`.
* Loop index `i = 1`: `arr[1] > largest` (32 > 55) is `False`.
* Loop index `i = 2`: `arr[2] > largest` (-97 > 55) is `False`.
* Loop index `i = 3`: `arr[3] > largest` (99 > 55) is `True`. `largest` becomes `99`.
* Loop index `i = 4`: `arr[4] > largest` (3 > 99) is `False`.
* Loop index `i = 5`: `arr[5] > largest` (67 > 99) is `False`.
* Returns `99`.

# Time Complexity

* **Time:** O(N)
* **Explanation:** The loop runs exactly `N` times to inspect each element once.

# Space Complexity

* **Space:** O(1)
* **Explanation:** No extra space proportional to input size is used.


```python
arr = [55,32,-97,99,3,67]
def large(arr):
    n = float("-inf") #indicating most smallest number
    largest = arr[0]
    for i in range(0,n):
        largest = max(largest,arr[i])
    return largest
large(arr)
```

# Problem Explanation

Finds the largest element in an array. However, this implementation contains a bug where the loop range boundary `n` is set to `-inf`, causing the loop to never execute.

# Approach Used

* **Technique:** Buggy Iterative Comparison.
* **Why it works:** The loop does not execute because `range(0, n)` where `n = -inf` evaluates to an empty range. Thus, it immediately returns the initial value of `largest`.
* **Key Idea:** The code attempts to loop from `0` to `-inf`, which represents an empty sequence, so no comparisons are performed.

# Dry Run

Input: `arr = [55, 32, -97, 99, 3, 67]`
* `n = -inf`, `largest = arr[0] = 55`
* Loop range is `range(0, -inf)`.
* Since the range is empty, the loop block is skipped.
* Returns `largest = 55`.

# Time Complexity

* **Time:** O(1)
* **Explanation:** No elements are checked in a loop; execution is constant time.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Constant auxiliary memory.
