```python
#method 1
arr = [55,32,-97,99,3,67]
def large(arr):
    n = len(arr)
    largest = arr[0]
    for i in range(0,n):
        largest = max(largest,arr[i])
    return largest
large(arr)
```

```python
#method 2
arr = [55,32,-97,99,3,67]
def large(arr):
    n = len(arr)
    largest = arr[0]
    for i in range(0,n):
        if arr[i]>largest:
            largest = arr[i]
    return largest
large(arr)
```

```python
arr = [55,32,-97,99,3,67]
def large(arr):
    n = float("-inf") #indicating most smallest number
    largest = arr[0]
    for i in range(0,n):
        largest = max(largest,arr[i])
    return largest
large(arr)
```
