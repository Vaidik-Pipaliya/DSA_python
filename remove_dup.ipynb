# brute-force

```python
num=[1,1,1,2,3,4,4,7,9,9,9,10]
def func(num):
    n=len(num)
    freq_map = {}
    for i in range(0,n):
        freq_map[num[i]] = 0
    j = 0
    for k in range(0,n):
        num[j]=k
        j+=1
    return j
```

# optimal

```python
num=[1,1,1,2,3,4,4,7,9,9,9,10]
def find(num):
    n = len(num)
    if n == 1:
        return 1
    i=0
    j=i+1
    while j<n:
        if num[j]!=num[i]:
            i+=1
            num[i],num[j] = num[j],num[i]
        j+=1
    return i+1 #total unique number jova mate
find(num)
```
