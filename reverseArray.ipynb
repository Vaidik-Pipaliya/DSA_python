# using recursion

```python
def func (arr,left,right):
    if left >= right:
        return
    arr[left],arr[right] = arr[right],arr[left]
    func(arr,left+1,right-1)

def rev(nums):  
    func(nums,0,len(nums)-1)
    return nums
nums = [5,9,8,2,3,0,6]  
print(rev(nums))

   
```
