# brute force

```python
nums=[3,9,5,6,7,2]
k = int(input("enter the numer for rotation"))
for _ in range(0,k):
    e  = nums.pop()
    nums.insert(0,e)
```

# Problem Explanation

Rotates an array to the right by `k` places. The user is prompted to enter the value of `k`.

# Approach Used

* **Technique:** Brute Force Array Shift.
* **Why it works:** In each of the `k` rotations, the last element of the list is popped and inserted at the beginning (index 0).
* **Key Idea:** Perform a single right shift `k` times.

# Dry Run

Input: `nums = [3, 9, 5]`, `k = 2`
* Loop run 1:
  * `e = nums.pop() = 5` -> `nums` becomes `[3, 9]`
  * `nums.insert(0, 5)` -> `nums` becomes `[5, 3, 9]`
* Loop run 2:
  * `e = nums.pop() = 9` -> `nums` becomes `[5, 3]`
  * `nums.insert(0, 9)` -> `nums` becomes `[9, 5, 3]`
* Final Output: `[9, 5, 3]`

# Time Complexity

* **Time:** O(k * N)
* **Explanation:** Popping from the end is O(1), but inserting at index 0 requires shifting all `N` elements which is O(N). This is repeated `k` times.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Operations are performed in-place.


# optimized  (k%n)

```python
def k_rotate(nums):
    k = int(input("enter the numer for rotation"))
    n = len(nums)
    rotation = k%n

    for _ in range(0,rotation):
        e = nums.pop()
        nums.insert(0,e)
    print (nums)
nums=[3,9,5,6,7,2]
k_rotate(nums)
```

# Problem Explanation

Rotates an array to the right by `k` places, optimized by using `k % n` to avoid redundant full-circle rotations.

# Approach Used

* **Technique:** Modulo-Optimized Shift.
* **Why it works:** Rotating a list of size `n` by `n` steps yields the same list. Taking `k % n` gives the minimal equivalent shifts needed.
* **Key Idea:** Eliminate redundant complete cycles of rotation.

# Dry Run

Input: `nums = [3, 9, 5]`, `k = 5`
* `n = 3`
* `rotation = 5 % 3 = 2`
* Loop run 1:
  * `e = nums.pop() = 5` -> `nums` becomes `[3, 9]`
  * `nums.insert(0, 5)` -> `nums` becomes `[5, 3, 9]`
* Loop run 2:
  * `e = nums.pop() = 9` -> `nums` becomes `[5, 3]`
  * `nums.insert(0, 9)` -> `nums` becomes `[9, 5, 3]`
* Final Output: `[9, 5, 3]`

# Time Complexity

* **Time:** O((k % N) * N)
* **Explanation:** The number of insert operations is reduced from `k` to at most `N - 1`. Each insert operation takes O(N) time.

# Space Complexity

* **Space:** O(1)
* **Explanation:** In-place modification of the array.
