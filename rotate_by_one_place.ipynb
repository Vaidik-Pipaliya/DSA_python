```python
def rotate(num):
    n = len(num)
    temp = num[n-1]
    for i in range(n-2,-1,-1):
        num[i+1] = num[i]
    num[0] = temp

num = [1,4,6,2,8,4,9]
rotate(num)
print(num)
```

# Problem Explanation

Rotates the elements of an array to the right by one position in-place.

# Approach Used

* **Technique:** In-place Shift.
* **Why it works:** It stores the last element in a temporary variable, shifts all other elements one index to the right starting from the back, and then assigns the temporary variable to index 0.
* **Key Idea:** Move each element `num[i]` to `num[i+1]`, wrapping the last element to the front.

# Dry Run

Input: `num = [1, 4, 6]`
* `n = 3`, `temp = num[2] = 6`
* Loop index `i` runs from `1` down to `0`:
  * `i = 1`: `num[2] = num[1]` -> `num` becomes `[1, 4, 4]`
  * `i = 0`: `num[1] = num[0]` -> `num` becomes `[1, 1, 4]`
* `num[0] = temp = 6` -> `num` becomes `[6, 1, 4]`

# Time Complexity

* **Time:** O(N)
* **Explanation:** The loop runs `N - 1` times to shift elements.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Only uses constant extra space for `temp` and loop indices.
