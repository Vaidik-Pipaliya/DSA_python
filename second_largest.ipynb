### built-in solution 

```python

nums=[55,32,97,-55,45,32,88,21]
nums.sort()
print(nums[-2])
```

# Problem Explanation

Finds the second largest element in an array by sorting the array and accessing the element at index `-2`.

# Approach Used

* **Technique:** Sorting.
* **Why it works:** Sorting the array arranges elements in ascending order. The largest element will be at index `-1` and the second largest will be at index `-2`.
* **Key Idea:** Sort and index from the end.

# Dry Run

Input: `nums = [55, 32, 97, -55, 45, 32, 88, 21]`
* Sort the list: `[-55, 21, 32, 32, 45, 55, 88, 97]`
* Print element at index `-2`: `88`

# Time Complexity

* **Time:** O(N log N)
* **Explanation:** Sorting takes O(N log N) time using Timsort.

# Space Complexity

* **Space:** O(N)
* **Explanation:** Python's sort method has a space complexity of O(N) in the worst case.


### better

```python
nums=[55,32,97,-55,45,32,88,21]
def second_large(nums):
    highest = float("-inf")
    sec_high =float("-inf")
    n = len(nums)
    for i in range(0,n):
        highest = max(highest,nums[i])
    for i in range(0,n):
        if nums[i]>sec_high and nums[i]!=highest:
            sec_high = nums[i]
    return sec_high
second_large(nums)
```

# Problem Explanation

Finds the second largest element in an array in two passes without sorting.

# Approach Used

* **Technique:** Two-Pass Maximum Search.
* **Why it works:** First pass finds the absolute largest element. Second pass finds the largest element that is strictly smaller than the absolute largest.
* **Key Idea:** Scan once for largest, then scan again for second largest excluding the largest.

# Dry Run

Input: `nums = [55, 32, 97]`
* `highest = -inf`, `sec_high = -inf`, `n = 3`
* First pass: Finds `highest = 97`.
* Second pass:
  * `i = 0`: `nums[0] = 55`. `55 > -inf` and `55 != 97` -> `sec_high = 55`
  * `i = 1`: `nums[1] = 32`. `32 > 55` (False)
  * `i = 2`: `nums[2] = 97`. `97 != 97` (False)
* Returns `55`.

# Time Complexity

* **Time:** O(N)
* **Explanation:** Performs two sequential linear scans of the array.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Constant auxiliary memory.


# optimized solution


```python
nums=[55,32,97,-55,45,32,88,21]
def sec_lar(nums):
    n = len(nums)
    largest = float("-inf")
    s_largest = float("-inf")
    for i in range(0,n):
        if nums[i]>largest:
            s_largest = largest
            largest = nums[i]
        elif nums[i]>s_largest and nums[i]!=largest:
            s_largest = nums[i]
    return s_largest
sec_lar(nums)

```

# Problem Explanation

Finds the second largest element in an array in a single pass.

# Approach Used

* **Technique:** Single-Pass Tracking.
* **Why it works:** It tracks the largest and second largest elements simultaneously. For each element, if it exceeds `largest`, `largest` is updated and the old `largest` is demoted to `s_largest`. If it is smaller than `largest` but larger than `s_largest`, only `s_largest` is updated.
* **Key Idea:** Keep a running track of both top-2 largest elements in one loop.

# Dry Run

Input: `nums = [55, 32, 97]`
* `n = 3`, `largest = -inf`, `s_largest = -inf`
* `i = 0` (55): `55 > -inf` -> `s_largest = -inf`, `largest = 55`
* `i = 1` (32): `32 > 55` (False). `32 > -inf` and `32 != 55` -> `s_largest = 32`
* `i = 2` (97): `97 > 55` -> `s_largest = 55`, `largest = 97`
* Returns `55`.

# Time Complexity

* **Time:** O(N)
* **Explanation:** Performs a single scan of the array.

# Space Complexity

* **Space:** O(1)
* **Explanation:** Constant auxiliary space.
